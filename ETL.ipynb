```python
import pandas as pd
```

```python
donaciones = pd.read_csv("data_temp/datos-donaciones.csv")
```

```python
donaciones = donaciones.rename(columns={"Unnamed: 0":"codigo", "NOMBRE.DEL.PROYECTO":"nombre_del_proyecto", "DONATARIO":"donatario", "DONANTE":"donante", "RUT":"rut", "Dígito":"digito", "MONTO.FONDO.MIXTO":"monto_fondo_mixto", "MONTO.INSTITUCIÓN":"monto_institucion", "monto.total.donacion.num":"monto_total_donacion_num", "Año":"año", "Mes":"mes", "Día":"dia"} )
# donaciones=donaciones.rename(columns={"Unnamed: 0":"donation_id", "NOMBRE.DEL.PROYECTO":"project_name", "DONATARIO":"donor", "DONANTE":"donator", "RUT":"id", "Dígito":"digit", "MONTO.FONDO.MIXTO":"mixed_fond_amount", "MONTO.INSTITUCIÓN":"institution_amount", "monto.total.donacion.num":"total_donation_amount_num", "Año":"year", "Mes":"month", "Día":"day"} )
```

```python
donaciones.head(3)

```

```python
donaciones["donatario"] = donaciones["donatario"].str.upper()
```

```python
donaciones.head(3)
```

```python
donaciones["mes"] = donaciones["mes"].astype(str).str.zfill(2)
```

```python
donaciones["dia"] = donaciones["dia"].astype(str).str.zfill(2)
```

```python
donaciones["fecha"] = donaciones["año"].astype(str) + "/" + donaciones["mes"].astype(str) +"/" + donaciones["dia"].astype(str)
```

```python
donaciones.head(3)
```

```python
donaciones["rut"] = donaciones["rut"].astype("str").str[0:7]
```

```python
donaciones=donaciones.drop(columns={"año", "mes", "dia"})
```

```python
donaciones["nombre_del_proyecto"] = donaciones["nombre_del_proyecto"].str.upper()
donaciones["donante"] = donaciones["donante"].str.upper()
```

```python
donaciones["monto_fondo_mixto"] = donaciones["monto_fondo_mixto"].fillna(0)
```

```python
donaciones
```

```python
def eliminarTildes(algo):
    algo=algo.str.replace(u"Á", "A")
    algo=algo.str.replace(u"É", "E")
    algo=algo.str.replace(u"Í", "I")
    algo=algo.str.replace(u"Ó", "O")
    algo=algo.str.replace(u"Ú", "U")   
    return algo
```

```python
donaciones["nombre_del_proyecto"] = eliminarTildes(donaciones["nombre_del_proyecto"])
donaciones["donatario"] = eliminarTildes(donaciones["donatario"])
donaciones["donante"] = eliminarTildes(donaciones["donante"])
```
